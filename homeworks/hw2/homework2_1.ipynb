# Solution

```python
import os
import re
import json
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.tools import tool
from langchain_core.messages import HumanMessage
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_mcp_adapters.client import MultiServerMCPClient
from langchain_core.output_parsers import StrOutputParser, JsonOutputParser
from langchain_core.messages import ToolMessage

# ---------------------------
# Initialize Gemini
# ---------------------------
llm = ChatGoogleGenerativeAI(
    model="gemini-2.0-flash",
    google_api_key=GEMINI_VERTEX_API_KEY,
    vertexai=True,
    temperature=0,
)
```

```python
@tool
def read_cv(file: str, text: str) -> str:
    """
    Reads and extracts the raw text content from a CV file.
    Always use this tool FIRST to get the candidate's background information before using any social media verification tools.

    Args:
        cv (dict): The CV file to read.
    """
    system_prompt = """You are an expert HR data extraction system. Your task is to extract highly accurate, structured information from the provided raw CV text.
    This extracted data will be used to verify the candidate's background against LinkedIn and Facebook databases.

    ### INSTRUCTIONS:
    1. Extract the candidate's full name, location, education history, work experience and other useful information.
    2. If a specific piece of information (like end year or location) is missing, output `null` for that field. Do not invent information.
    3. Normalize all dates to integer years (e.g., "Sept 2018 - Aug 2022" becomes start_year: 2018, end_year: 2022). If "Present" or "Current", set end_year to null.
    4. Normalize degrees to standard abbreviations if possible (e.g., "Bachelor of Science" -> "BSc", "Master" -> "MSc").
    5. You MUST output ONLY a valid JSON object. No extra text. No markdown. No explanation.

    ### OUTPUT JSON FORMAT:
    {{
      "name": "Full Name",
      "location": "City, Country (if available, else null)",
      "education": [
        {{
          "school": "University or School Name",
          "degree": "Degree type (e.g., BSc, MSc, PhD)",
          "field": "Major or Field of Study",
          "start_year": YYYY (integer),
          "end_year": YYYY (integer or null)
        }}
      ],
      "experience": [
        {{
          "company": "Company Name",
          "title": "Job Title",
          "start_year": YYYY (integer),
          "end_year": YYYY (integer or null)
        }}
      ],
      "skills": ["Skill 1", "Skill 2"]
    }}"""

    prompt = ChatPromptTemplate.from_messages([
        ('system', system_prompt),
        ('user', 'Here is the name of cv {file} and its content {text}')
    ])

    chain = prompt | llm | JsonOutputParser()
    return chain.invoke({'file': file, 'text': text})
```

```python
client = MultiServerMCPClient({
    "social_graph": {
        "transport": "http",
        "url": "https://ftec5660.ngrok.app/mcp",
        "headers": {"ngrok-skip-browser-warning": "true"}
    }
})

mcp_tools = await client.get_tools()
tools = mcp_tools + [read_cv]
```

```python
async def cv_verification_loop(cv, tools, max_turns: int = 30):
    """
    Iterative agentic loop for CV Verification using MCP tools.
    """
    tool_map = {tool.name: tool for tool in tools}
    main_agent_with_tools = llm.bind_tools(tools)

    system_prompt = """
    You are an expert HR AI Agent. Your task is to verify a candidate's CV against social media(Facebook and LinkedIn).

    YOUR MANDATORY WORKFLOW:
    1. **Extraction**: You need use the read_cv tool to extracte the information from cv.
    2. **LinkedIn Pass**: Use `search_linkedin_people` (with fuzzy=True if needed). Then use `get_linkedin_profile`.
    3. **Facebook Pass**: Use `search_facebook_users` and `get_facebook_profile` for secondary verification.
    4. **Fraud Detection**:
       - You can use `get_linkedin_interactions` to check for Ghost Accounts (post_count=0).
       - If references/connections are mentioned, you can check mutual friends using `get_facebook_mutual_friends`.

    SCORING CRITERIA:
    You must calculate a final `CONFIDENCE_SCORE` between 0.0 and 1.0.
    - If CONFIDENCE_SCORE > 0.5, the CV is considered VALID.
    - If CONFIDENCE_SCORE <= 0.5, the CV is considered INVALID.

    FINAL OUTPUT FORMAT:
    You MUST output your final decision exactly in this format:

    VERIFICATION_STATUS: [VALID or INVALID]
    CONFIDENCE_SCORE: [0.0 to 1.0]
    REASONING: [explaining why this score was given]
    DISCREPANCIES:
    - [Bullet points of differences found, if any]
    """
    file_name = cv['file']
    text = cv['text']
    history = [
        ("system", system_prompt),
        ("human", f"Please verify the candidate from this CV.\n\n File Name: {file_name}\n CV content:\n {text}")
    ]

    print(f"Starting verification...")

    for turn in range(max_turns):
      print(f"\n--- [Turn {turn+1}] Agent is thinking... ---")
      response = await main_agent_with_tools.ainvoke(history)
      history.append(response)
      tool_calls = getattr(response, "tool_calls", [])
      print(tool_calls)
      if not tool_calls:
        print(f"No more tool calls. Final report generated.")
        print("\n=== FINAL VERIFICATION REPORT ===")
        print(response.content)
        break

      for call in tool_calls:
        tool_name = call["name"]
        args = call.get("args", {})
        tool_call_id = call["id"]

        print(f"[Executing Tool] {tool_name}")
        if tool_name in tool_map:
          tool_result = await tool_map[tool_name].ainvoke(args)
        else:
          tool_result = {"error": f"Unknown tool: {tool_name}"}

        print(f"[Tool Result] Data fetched for {tool_name}.")

        history.append(ToolMessage(
            tool_call_id=tool_call_id,
            content=json.dumps(tool_result)
        ))

    return response.content
```

```python
def extract_score(text: str) -> float:
    pattern = r"CONFIDENCE_SCORE:\s*([0-9]*\.?[0-9]+)"
    match = re.search(pattern, text)

    if not match:
        raise ValueError("CONFIDENCE_SCORE not found in model output.")

    score = float(match.group(1))

    if score < 0.0:
        score = 0.0
    if score > 1.0:
        score = 1.0

    return score

async def evaluate_all_cvs(cv_list, tools: list) -> list[float]:
    score_list = []

    for i, cv_dict in enumerate(cv_list):
      cv_name = cv_dict.get('file', f'CV_{i+1}')
      print(f"\n{'='*20} Processing [{i+1}/{len(cv_list)}]: {cv_name} {'='*20}")

      report_text = await cv_verification_loop(cv_dict, tools)
      await asyncio.sleep(1.5)

      score = extract_score(report_text)
      score_list.append(score)

    return score_list
```

# Evaluation code

In the test phase, you will be given 5 CV files with fixed names:

    CV_1.pdf, CV_2.pdf, CV_3.pdf, CV_4.pdf, CV_5.pdf

Your system must process these CVs and output a list of 5 scores,
one score per CV, in the same order:

    scores = [s1, s2, s3, s4, s5]

Each score must be a float in the range [0, 1], representing the
reliability or confidence that the CV is valid (or meets the task criteria).

The ground-truth labels are binary:

    groundtruth = [0 or 1, ..., 0 or 1]

Each CV is evaluated independently using a threshold of 0.5:

- If score > 0.5 and groundtruth == 1 â†’ Full credit
- If score â‰¤ 0.5 and groundtruth == 0 â†’ Full credit
- Otherwise â†’ No credit

In other words, 0.5 is the decision threshold.

- Each CV contributes equally.
- Final score = (number of correct decisions) / 5


```python
# =====================================================
#  Evaluation code
# =====================================================

def evaluate(scores, groundtruth, threshold=0.5):
    """
    scores: list of floats in [0, 1], length = 5
    groundtruth: list of ints (0 or 1), length = 5
    """
    assert len(scores) == 5
    assert len(groundtruth) == 5

    correct = 0
    decisions = []

    for s, gt in zip(scores, groundtruth):
        pred = 1 if s > threshold else 0
        decisions.append(pred)
        if pred == gt:
            correct += 1

    final_score = correct / len(scores)

    return {
        "decisions": decisions,
        "correct": correct,
        "total": len(scores),
        "final_score": final_score
    }
```
